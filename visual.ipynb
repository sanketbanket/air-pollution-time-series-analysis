```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('F:\\dev_stuff\\DSP_TIME_SERIES\\air+quality\\AirQualityUCI.csv',sep = ';')


keys = df.keys()

for k in keys : 
    if k == "Date" or k == "Time": continue
    try : plt.plot([float(x.replace(',', '.')) for x in df[k].to_list()])
    except : plt.plot(df[k])
    plt.xlabel("Time")
    plt.ylabel("Reading")
    plt.title(k)
    plt.show()
```
